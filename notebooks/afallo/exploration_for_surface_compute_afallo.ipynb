# 1.Load .nii dataset 

```python
path = "/home/aurore/code/afallo/brain_tumor_detection_project/raw_data/segmentation/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData/BraTS-GLI-00000-000/BraTS-GLI-00000-000-t1c.nii.gz"
```

```python
import nibabel as nib

# Charger un fichier NIfTI
img = nib.load(path)

# Récupérer le header
hdr = img.header

# Exemple : afficher les dimensions
print("Dimensions (nx, ny, nz, nt):", hdr.get_data_shape())

# Exemple : taille des voxels (équivalent PixelSpacing en DICOM)
print("Voxel size (mm):", hdr.get_zooms())

# Exemple : matrice affine (orientation et position dans l’espace)
print("Affine matrix:\n", img.affine)

# Exemple : toutes les métadonnées disponibles
print(hdr)
```
